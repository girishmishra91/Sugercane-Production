```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1.Line Plot

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sine Wave')
plt.show()
```

#### 1.2)Custom Line Graph

```python
x=np.linspace(0,10,50)
y=np.sin(x)
plt.plot(x,y,"*",linestyle='--',color='green',markersize=10,alpha=.5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sine Wave')
plt.show()
```

##### 1.3)Line Plot with Multiple Lines

```python
x=np.linspace(0,10,50)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,"*",linestyle='--',color='green',markersize=10,alpha=.5,label="sin x")
plt.plot(x,y2,"*",linestyle='--',color='red',markersize=10,alpha=.5,label="cos x")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sine Wave')
plt.legend()
plt.show()
```

# 2.Scatter Plot

##### 2.1)Basic Scatter Plot

```python
x=np.random.rand(100)
y=np.random.rand(100)
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.show()
```

##### 2.2)Customize Scatter Plot

```python
x=np.random.rand(100)
y=np.random.rand(100)
z=np.random.rand(100)
plt.scatter(x,y,marker="*",c=z,alpha=.6,cmap='viridis')
plt.xlabel('x-axis')
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.colorbar()
plt.show()
```

##### 2.3)Scatter Plot with Multiple Classes

```python
x1=np.random.rand(500)-.5
y1=np.random.rand(500)-.5

x2=np.random.rand(500)+.5
y2=np.random.rand(500)+.5

plt.scatter(x1,y1,alpha=.6,marker='o')
plt.scatter(x2,y2,alpha=.6,marker='o')
plt.xlabel('x-axis')
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.colorbar()
plt.show()
```

##### 2.4)Hexbining Scatter Plot

```python
x=np.random.rand(20000)
y=np.random.rand(20000)
plt.hexbin(x,y,gridsize=20)
plt.xlabel('x-axis')
plt.ylabel("y-axis")
plt.title("Hexbining Plot")
plt.colorbar()
plt.show()
```

# 3.Bar Graph

#### 3.1) Basic bar graph

```python
x=np.array(['A','B','C','D'])
y=np.array([10,20,30,40])
plt.bar(x,y)
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.title("Bar Graph")
plt.show()
```

#### 3.2)Stacked Bar Graph

```python
data = np.random.rand(10,5)

plt.bar(np.arange(10), data[:,0] + data[:,1] + data[:,2] + data[:,3] + data[:,4], label = 'D-5')
plt.bar(np.arange(10), data[:,0] + data[:,1] + data[:,2] + data[:,3], label = 'D-4')
plt.bar(np.arange(10), data[:,0] + data[:,1] + data[:,2], label = 'D-3')
plt.bar(np.arange(10), data[:,0] + data[:,1],label = 'D-2')
plt.bar(np.arange(10), data[:,0], label = 'D-1')

plt.legend()
plt.show()
```

##### 3.3) Grouped Bar Graph

```python
data=np.array([[10,20,30],
               [40,50,60],
               [70,80,90]])
width=.25
x=np.arange(3)
plt.bar(x-width,data[:,0],width,label='D-1')
plt.bar(x,data[:,1],width,label="D-2")
plt.bar(x+width,data[:,2],width,label='D-3')
plt.legend()
plt.show()
```

# 4.Histogram

#### 4.1)Basic Histogram

```python
data=np.random.randn(1000)
plt.hist(data,bins=100)
plt.show()
```

##### 4.2)Cumulative Histogram

```python
data=np.random.randn(1000)
plt.hist(data,bins=10,cumulative=True)
plt.show()
```

##### 4.3)Histogram with Multiple Dataset

```python
data_1=np.random.randn(10000)
data_2=np.random.randn(10000)+4
plt.hist(data_1,bins=30,alpha=.7,label='D1')
plt.hist(data_2,bins=30,alpha=.7,label='D2')
plt.legend()
plt.show()
```

# 7.Piechart

```python
labels = ['Python', 'C++', 'Java', 'JavaScript', 'PHP']
sizes =  [21.3,        14.6 , 9.2,     8.5,       7.7]
explode = (0.13,        .1,    0,       0,           0)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']


plt.title("Programming Language Popularity")
plt.pie(sizes, colors=colors, labels=labels, autopct='%1.0f%%', shadow=True, explode=explode ,startangle=140)

plt.axis('equal')
plt.legend(title="Languages", loc="upper left")

plt.show()
```
